```r
library(dplyr) # importing required libraries 
library(rnaturalearth)
library(rnaturalearthdata)
library(ggplot2)
library(sf)
library(plotly)
library(ggthemes)
```

```r
suicides <- read.csv("https://raw.githubusercontent.com/popsnot/DATA201-Group-Project-2022/main/suicide-death-rates.csv") # reading csv directly from out github
head(suicides) # printing head of data
```

```r
names(suicides) <- c('Country', 'Code', 'Year', 'Suicides_Per100k') # renaming
head(suicides) # printing out head to see changes
```

```r
world <- ne_countries(scale = "medium", returnclass = "sf") # importing world data via rnaturalearth package
class(world) # shows class of data
```

```r
world = select(world, c(gu_a3, geometry)) # selecting chosen columns from new world data
```

```r
names(world) <- c('Code', 'Geometry') # merging by shared column
Suicides_geom = merge(suicides, world, by = "Code") %>% # merging world and suicides data by common key
  arrange(Year) # arranging by year

Suicides_geom # printing new dataframe
```

```r
Suicides_SF <- st_as_sf(Suicides_geom) # convert to sf object for plotting geometry
Suicides_SF <- Suicides_SF %>% 
  filter(Year == "2019" | Year == "1990") # filtering chosen years for comparison
```

```r
assoc_graph <- ggplot(data = Suicides_SF) + # first comparative plot
  geom_sf(aes(fill = Suicides_Per100k), 
          position = "identity") + 
  labs(fill='Suicides Per 100,000')  +
  scale_fill_viridis_c(option = "viridis") +
  facet_grid(.~Year) # separate plot for each year
  
assoc_graph + theme_map() + theme(legend.position = "bottom") # printing plot with theme_map() and adding legend position
```

```r
Suicides_Countries <- Suicides_SF %>% # filter for chosen countries
  filter(Country == "New Zealand" | Country == "South Korea" | Country == "Netherlands" | Country == "United States")
```

```r
ggplot(Suicides_Countries,                                      # Grouped barplot using ggplot2
       aes(x = factor(Year),
           y = Suicides_Per100k,
           fill = Country)) +
  geom_bar(stat = "identity", # adding bars to represent data
           position = "dodge")
```

```r
Suicides_Countries %>%
  ggplot( aes(factor(Year), Suicides_Per100k, color=Country, label= Code)) +
  geom_point(size = 6)+ # adding points to represent data
  labs(title = "Suicides Per 100,000 By Year", x = "Year", y = "Suicide Rates Per 100,000") # adding labels
```
