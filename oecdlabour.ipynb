```r
oecdlabour_df <- read.csv('https://stats.oecd.org/sdmx-json/data/DP_LIVE/.HRWKD.../OECD?contentType=csv&detail=code&separator=comma&csv-lang=en') #reads the file and assigns it to the variable bike_df
glimpse(oecdlabour_df) #the glimpse function allows us to have a peek at what the dataset will look like
```

```r
oecdlabour_df
```

```r
oecdlabour_df <- (oecdlabour_df %>% select(LOCATION, TIME, Value))
oecdlabour_df
```

```r
median_hrs <- (oecdlabour_df %>% group_by(LOCATION) %>%
    summarise(median_value = median(Value)))
median_hrs
```

```r
filtered_oecd <- oecdlabour_df[oecdlabour_df$LOCATION %in% c('ISR', 'NLD', 'NZL', 'KOR', 'USA'),]
filtered_oecd
```

```r
filtered_oecd %>% 
    ggplot(aes(x = TIME, y = Value, col=LOCATION, size=10)) + 
    geom_freqpoly(stat='identity') + #makes a barplot and makes the graph look nice
    labs(x = 'year', y = 'Hours Worked', title = '') #labels the graph accurately 
```
